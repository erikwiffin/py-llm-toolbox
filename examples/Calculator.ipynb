```python
from openai import OpenAI

client = OpenAI(base_url="http://localhost:1234/v1", api_key="not-needed")
```

```python
import json
from toolbox import Toolbox
from numbers import Number

# 1. Setup Toolbox
toolbox = Toolbox()

# 2. Register Functions
@toolbox.function(description="A function that performs an addition operation on two numbers")
@toolbox.parameter(name="a", description="The first number to add")
@toolbox.parameter(name="b", description="The second number to add")
def add(a: Number, b: Number):
    return a + b

@toolbox.function(description="A function that performs a subtraction operation on two numbers")
@toolbox.parameter(name="a", type="number", description="The number to subtract from")
@toolbox.parameter(name="b", type="number", description="The number to subtract")
def subtract(a, b):
    return a - b

print(json.dumps(toolbox.tools, indent=2))
```

```python
messages = [
    {"role": "system", "content": "You are a very fancy calculator."},
    {"role": "user", "content": "What is 1234 + 5678?"},
]

response = client.chat.completions.create(
    model="mistral-small-3.1-24b-instruct-2503",
    messages=messages,
    tools=toolbox.tools
)

output = toolbox.execute(response.choices[0].message.tool_calls)

response = client.chat.completions.create(
    model="mistral-small-3.1-24b-instruct-2503",
    messages=[
        *messages,
        *output,
    ],
    tools=toolbox.tools
)

print(response.choices[0].message.content)
```
